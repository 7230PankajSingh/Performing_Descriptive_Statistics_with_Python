7. Calculate the z score of the following numbers:

X = [4.5,6.2,7.3,9.1,10.4,11]

```python
import numpy as np

# Data
X = np.array([4.5, 6.2, 7.3, 9.1, 10.4, 11])

# Calculate mean and standard deviation
mean_X = np.mean(X)
std_X = np.std(X)

# Calculate Z-scores
z_scores = (X - mean_X) / std_X

print("Z-scores:", z_scores)
```

```python
# Identify outliers using Z-score
outliers = (df_bank['Z_Income'] < -3) | (df_bank['Z_Income'] > 3)
outliers = (df_bank['Z_Income'] < -3) | (df_bank['Z_Income'] > 3)

# Count the number of outliers
num_outliers = outliers.sum()  # `sum()` counts the number of True values
print(num_outliers)
print(df_bank['outliers'])
```

```python
# Data
df_Income = df_bank['Income']

# Calculate mean and standard deviation
mean_I = df_Income.mean()
std_I = df_Income.std()

# Calculate Z-scores and add it to the DataFrame
df_bank['Z_Income'] = (df_Income - mean_I) / std_I

# Identify outliers using Z-scores
outliers = (df_bank['Z_Income'] < -3) | (df_bank['Z_Income'] > 3)

# Count the number of outliers
num_outliers = outliers.sum()  # `sum()` counts the number of True values
print(f"Number of observations outside ¬±3œÉ: {num_outliers}")

# Optional: Display the rows that are outliers
print("Outliers:")
print(df_bank[outliers])
```
